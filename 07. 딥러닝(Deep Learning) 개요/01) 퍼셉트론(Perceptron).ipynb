# 2.   (Single-Layer Perceptron)

-    ( : input layer)      (output layer)  



- (gate) :     0 1     

   AND, NAND, OR   , XOR   .

- AND  :     1   1 

```python
def AND_gate(x1, x2):
    w1=0.5
    w2=0.5
    b=-0.7
    result=w1*x1+w2*x2+b
    if result<0:
        return 0
    else:
        return 1
```

AND        1  1 

```python
AND_gate(0, 0), AND_gate(0, 1), AND_gate(1, 0), AND_gate(1, 1)
```

- NAND  :     1   0 

```python
def NAND_gate(x1, x2):
    w1=-0.5
    w2=-0.5
    b=0.7
    result=w1*x1+w2*x2+b
    if result<=0:
        return 0
    else:
        return 1
```

```python
NAND_gate(0, 0), NAND_gate(0, 1), NAND_gate(1, 0), NAND_gate(1, 1)
```

- OR  :     0   0 

```python
def OR_gate(x1, x2):
    w1=0.6
    w2=0.6
    b=-0.5
    result=w1*x1+w2*x2+b
    if result<=0:
        return 0
    else:
        return 1
```

```python
OR_gate(0, 0), OR_gate(0, 1), OR_gate(1, 0), OR_gate(1, 1)
```
