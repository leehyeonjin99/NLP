# 1.  퍼셉트론(Perceptron)

- 초기의 형태의 인공 신경망
- 다수의 입력으로부터 하나의 결과를 내보내는 알고리즘
- 신경 세포 뉴런의 동작과 유사


- 다수의 입력을 받는 퍼셉트론의 그림
  

x : 입력값, w : 가중치(weight), y : 출력값  

가중치의 값이 크다. → 해당 값이 중요하다.

각 입력값이 가중치와 곱해져서 인공 뉴련에 보내진다.  
→ 각 입력값과 그에 해당하는 가중치의 곱의 전체 합이 임계치를 넘으면 종착지에 있는 인공 뉴련은 출력 신호로서 1을 출력하고, 그렇지 않은 경우에는 0을 출력한다.  
→ 계단 함수(Step Function)





임계치를 좌변으로 넘기고 편향 b(bias)로 표현 가능





- 활성화 함수(Activation Function) : 뉴런에서 출력값을 변경시키는 함수

# 2. 단층  퍼셉트론(Single-Layer Perceptron)

- 값을 보내는 단계 (입력층 : input layer) 와 값을 받아서 출력하는 단계 (output layer) 로만 이루어짐



- 게이트(gate) : 컴퓨터에서 두 개의 값 0과 1을 입력받아 하나의 값으로 출력하는 회로

단층 퍼셉트론을 이용하면 AND, NAND, OR 게이트는 구현 가능하지만, XOR 게이터는 구현 불가능하다.

- AND 게이트 : 두 개의 입력값이 모두 1인 경우에만 출력값이 1인 구조

```python
def AND_gate(x1, x2):
    w1=0.5
    w2=0.5
    b=-0.7
    result=w1*x1+w2*x2+b
    if result<0:
        return 0
    else:
        return 1
```

AND 게이터의 입력값을 모두 넣어보면 오직 두개의 입력값이 1인 경우에만 1을 출력

```python
AND_gate(0, 0), AND_gate(0, 1), AND_gate(1, 0), AND_gate(1, 1)
```

- NAND 게이트 : 두 개의 입력값이 모두 1인 경우에만 출력값이 0인 구조

```python
def NAND_gate(x1, x2):
    w1=-0.5
    w2=-0.5
    b=0.7
    result=w1*x1+w2*x2+b
    if result<=0:
        return 0
    else:
        return 1
```

```python
NAND_gate(0, 0), NAND_gate(0, 1), NAND_gate(1, 0), NAND_gate(1, 1)
```

- OR 게이트 : 두 개의 입력값이 모두 0인 경우에만 출력값이 0인 구조

```python
def OR_gate(x1, x2):
    w1=0.6
    w2=0.6
    b=-0.5
    result=w1*x1+w2*x2+b
    if result<=0:
        return 0
    else:
        return 1
```

```python
OR_gate(0, 0), OR_gate(0, 1), OR_gate(1, 0), OR_gate(1, 1)
```

AND, NAND, OR 게이트는 시각화 하였을 때, 출력값에 따라 하나의 직선으로 나누는 것이 가능하다.



- XOR 게이트 : 입력값 두 개가 서로 다른 값을 갖고 있을 때에만 출력값이 1인 구조
    


# 3. 다층 퍼셉트론 (Multi-Layer Perceptron, MLP)

- 단층 퍼셉트론의 중간에 층을 하나 이상 더 추가
- **은닉층** : 입력층과 출력층 사이에 존재하는 층



- **심층 신경망(Deep Neural Network, DNN)** : 은닉층이 2개 이상인 신경망
